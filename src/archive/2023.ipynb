```python
from country_to_code import country_to_code
```

```python
f = open('2023.html', 'r')
lines = f.readlines()
```

```python
def extract(line):
    return line[line.index('">')+2:line.index('</')]

n = 0
inc = 1
lasttotal = ""
for i, line in enumerate(lines):
    if line != '<tr>\n' or 'Last Name' in lines[i+1]:
        continue
    first = extract(lines[i + 2])
    last = extract(lines[i + 1])
    last = " ".join(s[0] + s[1:].lower() for s in last.split(" "))
    code = extract(lines[i + 3])
    if code in country_to_code:
        code = country_to_code[code]
    theory = extract(lines[i + 5])
    experiment = extract(lines[i + 4])
    total = extract(lines[i + 6])
    medal = {"Gold":"G","Silver":"S","Bronze":"B","HM":"H"}[extract(lines[i + 7])]
    if total == lasttotal:
        inc += 1
    else:
        n += inc
        inc = 1
    lasttotal = total
    print(f"2023,{n},{first} {last},{code},{medal},{theory},{experiment},{total},")
```
