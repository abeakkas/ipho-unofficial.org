```python
def raw_to_row(raw):
    raw = raw.strip()
    raw = raw.replace(") ", ")\n")
    raw = raw.split("\n")
    rows = []
    for g in raw:
        name, country = g.split("(")
        name = name.strip()
        name = name.split(" ")
        while name[0].isupper():
            lastname = name[0][0] + name[0][1:].lower()
            name = name[1:] + [lastname]
        name = " ".join(name)
        country = country.replace(")", "")
        code = country_to_code[country]
        rows.append([name, code])
    return rows
```
